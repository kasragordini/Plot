```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
```

```python

# Load your datasets (assuming they are in CSV format, you can adjust this part to your actual data source)
# Replace this with the path to your actual datasets
train_df = pd.read_excel("C:\\Users\\Kasra\\Desktop\\Pooya\\Mmax_train_data.xlsx")  # Assuming it has columns: 'Observed', 'Predicted'
test_df = pd.read_excel("C:\\Users\\Kasra\\Desktop\\Pooya\\Mmax_test_data.xlsx")      # Assuming it has columns: 'Observed', 'Predicted'

# Check the column names in your datasets
print("Train Data Columns: ", train_df.columns)
print("Test Data Columns: ", test_df.columns)
```

```python
# Calculate R2, MSE, and RMSE for the test data
r2_train = r2_score(train_df['Observed'], train_df['Predicted'])
mse_train = mean_squared_error(train_df['Observed'], train_df['Predicted'])
rmse_train = np.sqrt(mse_train)

# Print the results for training data
print('Training Data:')
print(f'R²: {r2_train}')
print(f'MSE: {mse_train}')
print(f'RMSE: {rmse_train}')
```

```python
# Calculate R2, MSE, and RMSE for the test data
r2_test = r2_score(test_df['Observed'], test_df['Predicted'])
mse_test = mean_squared_error(test_df['Observed'], test_df['Predicted'])
rmse_test = np.sqrt(mse_test)

# Print the results for training data
print('Testing Data:')
print(f'R²: {r2_test}')
print(f'MSE: {mse_test}')
print(f'RMSE: {rmse_test}')
```

```python
# Scatter plot for training data
plt.figure(figsize=(10, 6))
plt.scatter(train_df['Observed'], train_df['Predicted'], color="blue", label="Training Data")
plt.plot([train_df['Observed'].min(), train_df['Observed'].max()], [train_df['Observed'].min(), train_df['Observed'].max()], 'k--', lw=2)
plt.title('Observed vs Predicted - Training Data')
plt.xlabel('Observed')
plt.ylabel('Predicted')
plt.legend()
plt.show()
```

```python

# Scatter plot for test data
plt.figure(figsize=(10, 6))
plt.scatter(test_df['Observed'], test_df['Predicted'], color="green", label="Test Data")
plt.plot([test_df['Observed'].min(), test_df['Observed'].max()], [test_df['Observed'].min(), test_df['Observed'].max()], 'k--', lw=2)
plt.title('Observed vs Predicted - Test Data')
plt.xlabel('Observed')
plt.ylabel('Predicted')
plt.legend()
plt.show()
```

```python
# Create subplots for better visualization
plt.figure(figsize=(12, 6))

# Training data regression and scatter plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
sns.regplot(x='Observed', y='Predicted', data=train_df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Training Data: Observed vs Predicted')
plt.xlabel('Observed')
plt.ylabel('Predicted')

# Test data regression and scatter plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
sns.regplot(x='Observed', y='Predicted', data=test_df, scatter_kws={'color': 'magenta'}, line_kws={'color': 'black'})
plt.title('Test Data: Observed vs Predicted')
plt.xlabel('Observed')
plt.ylabel('Predicted')

# Show the plots
plt.tight_layout()
plt.show()
```
