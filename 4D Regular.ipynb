```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Load your dataset
# Assuming the file has headers in the format of 'Design Scenario', 'CVaR 1/FS', 'CVaR_Hmax', 'CVaR_Mmax', 'CVaR_Vmax'
file_path = "C:\\Users\\Kasra\\Desktop\\Pooya\\New folder\\4D - regular.xlsx"  # Adjust path accordingly
df = pd.read_excel(file_path)
```

```python
# Extract the data from your DataFrame
x = df['CVaR FS']
y = df['CVaR Hmax (mm)']
z = df['CVaR Mmax (kN.m)']
color = df['CVaR Vmax (kN)']  # Fourth dimension represented as color
```

```python
# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Use color or size for the fourth dimension
scat = ax.scatter(x, y, z, c=color, cmap='viridis', s=100)  # You can adjust the 's' for size

# Add labels
ax.set_xlabel('CVaR FS')
ax.set_ylabel('CVaR Hmax (mm)')
ax.set_zlabel('CVaR Mmax (kN.m)')

# # Add a color bar to show the values of the fourth dimension (CVaR_Vmax)
# plt.colorbar(scat, label='CVaR_Vmax')

# Create colorbar and adjust its position
cbar = plt.colorbar(scat, ax=ax, pad=0.15)  # Use 'pad' to move the colorbar
cbar.set_label('CVaR Vmax (kN)')

plt.show()
```

```python
# Create a 2D scatter plot using 'CVaR 1/FS' and 'CVaR_Hmax'
fig, ax = plt.subplots()

# Plot the 2D scatter with the same color mapping as the 4D plot
scat_2d = ax.scatter(x, y, c=color, cmap='viridis', s=100)  # 'x' is 'CVaR 1/FS', 'y' is 'CVaR_Hmax'

# Add labels
ax.set_xlabel('CVaR FS')
ax.set_ylabel('CVaR Hmax (mm)')

# Use the same colorbar as the 4D plot
cbar = plt.colorbar(scat_2d, ax=ax)
cbar.set_label('CVaR Vmax (kN)')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a 2D scatter plot using 'CVaR 1/FS' and 'CVaR_Hmax'
fig, ax = plt.subplots()

# Plot the 2D scatter with the same color mapping as the 4D plot
scat_2d = ax.scatter(x, z, c=color, cmap='viridis', s=100)  # 'x' is 'CVaR 1/FS', 'y' is 'CVaR_Hmax'

# Add labels
ax.set_xlabel('CVaR FS')
ax.set_ylabel('CVaR Mmax (kN.m)')

# Use the same colorbar as the 4D plot
cbar = plt.colorbar(scat_2d, ax=ax)
cbar.set_label('CVaR Vmax (kN)')

plt.show()
```

```python
# Create a 2D scatter plot using 'CVaR 1/FS' and 'CVaR_Vmax'
fig, ax = plt.subplots()

# Normalize 'CVaR_Hmax' to use as transparency (alpha)
normalized_alpha = (y - y.min()) / (y.max() - y.min())  # Normalize values between 0 and 1

# Use transparency (alpha) to represent 'CVaR_Hmax'
scat_2d = ax.scatter(x, color, c=color, cmap='viridis', alpha=normalized_alpha)  # 'x' is 'CVaR 1/FS', 'color' is 'CVaR_Vmax'

# Add labels
ax.set_xlabel('CVaR FS')
ax.set_ylabel('CVaR Vmax (kN)')

# Use the same colorbar as the 4D plot to show 'CVaR_Vmax'
cbar = plt.colorbar(scat_2d, ax=ax)
cbar.set_label('CVaR Vmax (kN)')

plt.show()
```

```python
# Example data (replace with your actual data)
x = np.random.rand(20)  # CVaR 1/FS
vmax = np.random.rand(20) * 1000  # CVaR_Vmax

# Create a 4x5 grid of subplots (20 total plots)
fig, axes = plt.subplots(4, 5, figsize=(15, 10))

# Loop through each data point and create a small plot
for i, ax in enumerate(axes.flat):
    ax.scatter(x[i], vmax[i], s=100, c='blue')  # Scatter plot for each data point
    ax.set_xlabel('CVaR 1/FS')
    ax.set_ylabel('CVaR_Vmax')
    ax.set_title(f'Design Scenario {i+1}')
    
    # Adjust x and y axis limits for consistency across subplots
    ax.set_xlim(x.min() - 0.1, x.max() + 0.1)
    ax.set_ylim(vmax.min() - 100, vmax.max() + 100)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
